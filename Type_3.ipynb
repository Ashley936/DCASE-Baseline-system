```python
import os
import pandas as pd
import numpy as np
import librosa
import tensorflow as tf
```

```python
# Load annotations file
annotations = pd.read_csv('output.csv')

# Define audio settings
sr = 44100
duration = 1.0
hop_length = 512
n_mels = 128
all_keys=['people walking', 'car', 'large vehicle', 'brakes squeaking', 'people speaking', 'children']
```

```python
# Define empty lists for features and labels
segment_features = []
segment_labels = []
data_dir = "./applications/data/TUT-sound-events-2017-development"
```

```python
# Extract features and labels from audio segments
for index, row in annotations.iterrows():
    file_path = os.path.join(data_dir, row['filename'])
    audio, _ = librosa.load(file_path, sr=sr)

    # Split the segment into frames
    frames = librosa.util.frame(audio, frame_length=int(duration * sr), hop_length=int(duration * sr))

    for i in range(frames.shape[1]):
        segment = frames[:, i]

        # Extract mel spectrogram
        mel_spec = librosa.feature.melspectrogram(segment, sr=sr, n_mels=n_mels, hop_length=hop_length)
        log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)

        # Append features and labels to the lists
        segment_features.append(log_mel_spec)
        label = np.zeros(6)
        label[all_keys.index(row['event_type'])] = 1
        segment_labels.append(label)
```
